```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
data_path = "./data/data_agg.csv"

df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])
```

```python
df['post_treatment'] = df['year'] == 2025
```

```python
df
```

```python
df.columns
```

```python
model = smf.ols(
    'estimated_average_ridership ~ post_treatment + C(month) + C(day_of_week) + C(hour_of_day) + C(origin_nta2020)',
    data=df
).fit(cov_type='cluster', cov_kwds={'groups': df['origin_nta2020']})

print(model.summary())
```
