```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_path = "./data/data_agg.csv"

df = pd.read_csv(data_path)
```

```python
# 2. Load NYC NTA2020 shapefile (publicly available)
nta = gpd.read_file('./data/nynta2020_25c/nynta2020.shp')
nta
```

```python
# 3. Merge your data with the geometry
merged = nta.merge(df, left_on="NTA2020", right_on="origin_nta2020", how="left")
```

```python
merged
```

```python
# 4. Plot the map
fig, ax = plt.subplots(figsize=(10, 10))
merged.plot(
    column="estimated_average_ridership",  # e.g., number of events or rate
    ax=ax,
    legend=True,
    cmap="OrRd",  # red color scale
    missing_kwds={"color": "lightgrey"}
)
plt.title("Events by NYC NTA (2020)")
plt.axis("off")
plt.show()
```
